#  <span style="color:#2E86C1;">Data Types</span>

## <span style="color:#117A65;">What is Data Type?</span>

A data type is like a label that tells the computer what kind of information you’re storing and what you can do with it.

## <span style="color:#117A65;">Why we need Data Types:</span>

- <span style="color:#566573;">To store different kinds of information.</span>  
- <span style="color:#566573;">To do right operations.</span>  
- <span style="color:#566573;">To save computer memory</span>

# <span style="color:#2E86C1;">Python Data Types Categories</span>

## <span style="color:#A04000;">Numerical Types:</span>

- <b style="color:#2471A3;">List</b>: Ordered, mutable collection that allows objects.  
    - <span style="color:#566573;">25, -17, 100000000</span>  
- <b style="color:#2471A3;">Float</b>: Number with decimal points.  
    - <span style="color:#566573;">17.77, -2.99, 1.5e10</span>  
- <b style="color:#2471A3;">Complex</b>: Number with real and imaginary parts.  
    - <span style="color:#566573;">5 + 6j</span>  

## <span style="color:#A04000;">Text Type:</span>

- <b style="color:#2471A3;">String(str)</b>: Sequence of character enclosed in quotes.  
    - <span style="color:#566573;">“Hello”</span>  

## <span style="color:#A04000;">Boolean Type:</span>

- <b style="color:#2471A3;">Boolean(bool)</b>: Represents True or False values.  
    - <span style="color:#566573;">True, False</span>  

## <span style="color:#A04000;">Sequence Types:</span>

- <b style="color:#2471A3;">List</b>: Ordered, mutable collection that allows objects.  
    - <span style="color:#566573;">[“apple”, “banana”, 1, 2]</span>  
- <b style="color:#2471A3;">Tuple</b>: Ordered, immutable collection that allows objects.  
    - <span style="color:#566573;">(“red”, “green”, 3, 4)</span>  
- <b style="color:#2471A3;">Range</b>: Represents a sequence of numbers, commonly used in loop.  
    - <span style="color:#566573;">range(6)  # 0, 1, 2, 3, 4, 5</span>  

## <span style="color:#A04000;">Mapping Type:</span>

- <b style="color:#2471A3;">Dictionary(dict)</b>: Unordered collection of key-value pairs  
    - <span style="color:#566573;">{“name”: “CSE”, “student”: 40, “city”: “Saidpur”}</span>  

## <span style="color:#A04000;">Set Types:</span>

- <b style="color:#2471A3;">Set</b>: Unordered collection of unique elements.  
    - <span style="color:#566573;">{“red”, “green”, 1, 2}</span>  
- <b style="color:#2471A3;">Frozenset</b>: Immutable version of a set.  
    - <span style="color:#566573;">frozenset({“red”, “green”, “blue”})</span>  

## <span style="color:#A04000;">Special Type:</span>

- <b style="color:#2471A3;">NoneType</b>: Represent the absence of a value.  

## <span style="color:#A04000;">User-Defined Type:</span>

- <b style="color:#2471A3;">Class</b>: Obj blueprint.
